```python
import os

try:
    from google.colab import files
except ImportError:
    print("❌ HATA: Bu kod sadece Google Colab sitesinde çalışır!")
    print("Lütfen bu dosyayı https://colab.research.google.com/drive/1XDqFUsKTALWZz4UCdnwJyU9n2uM3g94l adresine yükleyin.")
    files = None

if files:
    print("Kontrol Ediliyor...")
    apk_file = None

    # APK dosyasını bul ve indir
    if os.path.exists('bin'):
        for file in os.listdir('bin'):
            if file.endswith('.apk'):
                apk_file = os.path.join('bin', file)
                print(f"✅ BAŞARILI! APK Bulundu: {apk_file}")
                files.download(apk_file)
                break

    if not apk_file:
        print("❌ HATA: 'bin' klasörü boş veya oluşturulamadı! (Lütfen yukarıdaki kırmızı hataları kontrol edin)")
```
